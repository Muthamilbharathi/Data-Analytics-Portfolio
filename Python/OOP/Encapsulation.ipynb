```python
class employee:
    def __init__ (self):
        self.__salary=-10000


    def updatedsalary(self,salary):
        if salary>0:
            self.__salary=salary
            return "updatedsalary"
        else:
            return "Invalid salary"
    def get_salary(self):
        return self.__salary

e1=employee()
e1.updatedsalary(100)

print(e1.get_salary())


```
