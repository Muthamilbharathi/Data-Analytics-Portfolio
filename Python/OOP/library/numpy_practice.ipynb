```python
call_duration=np.array([5,7,10,15,6,9,12,8])
satisfacation=np.array([5,4,3,5,2,4,5,3])
calls=np.array([120,130,160,180,150,170,140,200])

#average call duration
print(np.average(call_duration))
#maximum satisfacation
print(np.max(satisfacation))
#calls greater than 150
print(calls[np.where(calls>150)])
#sort call duration
np.sort(call_duration)
#satisfacation is 5
np.where(satisfacation==5)
```
