```python
from abc import ABC,abstractmethod
class employee(ABC):
    @abstractmethod
    def work(self):
        pass
class dev(employee):
         def work(self):
               return"Developing"
a=dev()  
print(a.work())

class HR(employee):
    def work(self):
        pass
      
print(HR().work())

class DA(employee):
     def work(self):
          return"Analysing"

DA().work()
```
